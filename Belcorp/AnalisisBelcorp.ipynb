# Analisis estadistico de los dummy datasets de Belcorp

## Analisis de clientes

```python
import pandas as pd
import matplotlib
```

```python
clientes = pd.read_json('https://belc-hackathon2023.s3.amazonaws.com/clients.json')
clientes
```

```python

# Resumen estadístico de datos numéricos
clientes.describe()
```

```python

# Resumen de datos categóricos
clientes.describe(include=['object'])
```

```python
# Cantidad de clientes por país
clientes_por_pais = clientes['cod_pais'].value_counts()

clientes_por_pais
```

```python
# Calcular la antigüedad del cliente (en días) desde la creación hasta la última modificación
clientes['antiguedad'] = (pd.to_datetime(clientes['fec_modificacion']) - pd.to_datetime(clientes['fec_creacion'])).dt.days
clientes['antiguedad'].describe()
```

## Analisis de Consultoras

```python
consultoras = pd.read_json('https://belc-hackathon2023.s3.amazonaws.com/consultants.json')
consultoras
```

```python
# Resumen estadístico de datos numéricos
consultoras.describe()
```

```python
# Resumen de datos categóricos
consultoras.describe(include=['object'])
```

```python
from datetime import datetime

# Convertir fechanacimiento a tipo datetime y calcular la edad
consultoras['fechanacimiento'] = pd.to_datetime(consultoras['fechanacimiento'])
consultoras['edad'] = (datetime.now() - consultoras['fechanacimiento']).dt.days // 365

# Mostrar el DataFrame completo con la columna de edad
consultoras.describe()
#Edad mas alta de 25, promedio 40 años
```

```python

# Calcular antigüedad desde el registro
consultoras['fecharegistro'] = pd.to_datetime(consultoras['fecharegistro'])
consultoras['antiguedad_dias'] = (datetime.now() - consultoras['fecharegistro']).dt.days
consultoras.describe()
#Promedio de antiguedad 18 meses
```

```python

# Distribución de consultoras por país
consultoras_por_pais = consultoras['codpais'].value_counts()
consultoras_por_pais
#Colombia es la que tiene mas consultoras
#2 peru
#3 mexico
```
