## Ejercicio 1: Cargar los datos

1. Cargue la base de datos de trata de personas y tráfico de migrantes desde la siguiente fuente de información [datos.gov.co](https://www.datos.gov.co/Seguridad-y-Defensa/TRATA-DE-PERSONAS/95c7-mm6s/about_data) en un DataFrame de pandas.


```python
#La base de datos de trata y trafico la extraemos desde la api del gobierno con el link: https://www.datos.gov.co/resource/95c7-mm6s.csv?$limit=1747
import pandas as pd #Importación de la libreria pandas
url_data = 'https://www.datos.gov.co/resource/95c7-mm6s.csv?$limit=1747' #Usamos como limite 1747 ya que es el numero maximo de datos
data = pd.read_csv(url_data)
len(data) #Confirmamos que nuestros datos llegaron completos
```

```python
#Pasar los datos a un dataframe de pandas
df = pd.DataFrame(data)
df
```

```python
#Identifica cuál es el municipio con más casos reportados de trata de personas.
# Filtrar los casos de trata de personas y agrupar por municipio, sumando la cantidad de casos
municipio_max_trata = df_clean[df_clean['descripcion_conducta'] == "articulo 188 a.- trata de personas"].groupby('municipio')['cantidad'].sum()

# Mostrar el resultado
print(f"El municipio con más casos reportados de trata de personas es {municipio_max_trata.idxmax()} con {municipio_max_trata.max()} casos.")
```

## Ejercicio 5: Análisis temporal de casos

1. Convierte la columna `fecha_hecho` al tipo de dato `datetime`.
2. Crea una nueva columna que extraiga el año de los hechos.
3. Agrupa los casos por año y genera un gráfico de línea que muestre la evolución de la cantidad de víctimas a lo largo del tiempo.



```python
#Convertir la columna fecha_hecho a datetime
df_clean['fecha_hecho'] = pd.to_datetime(df_clean['fecha_hecho'])
```

```python
#Extraer el año de los hechos y añadirlo como una nueva columna
df_clean['año_hecho'] =  df_clean['fecha_hecho'].dt.year
df_clean
```

```python
#Agrupa los casos por año y genera un gráfico de línea que muestre la evolución de la cantidad de víctimas a lo largo del tiempo.
import matplotlib.pyplot as plt

# Agrupar los datos por año y sumar la cantidad de víctimas
victimas_por_año = df_clean.groupby('año_hecho')['cantidad'].sum()

# Crear el gráfico de línea
plt.figure(figsize=(10,6))
plt.plot(victimas_por_año.index, victimas_por_año.values, marker='o', linestyle='-', color='b')

# Añadir títulos y etiquetas
plt.title('Evolución de la cantidad de víctimas a lo largo del tiempo', fontsize=14)
plt.xlabel('Año', fontsize=12)
plt.ylabel('Cantidad de víctimas', fontsize=12)

# Mostrar el gráfico
plt.grid(True)
plt.show()
```

## Ejercicio 6: Visualización de datos con matplotlib

1. Crea un gráfico de barras que muestre la cantidad total de víctimas por delito por cada año.
2. Genera un gráfico de barras que muestre la cantidad de víctimas en los 5 departamentos con más casos reportados para el año 2024.
3. Haz un gráfico de torta que muestre la proporción de casos de trata de personas frente a tráfico de migrantes del año 2024.




```python
#gráfico de barras que muestre la cantidad total de víctimas por delito por cada año.
# Agrupar los datos por año y delito
delitos_por_año = df_clean.groupby(['año_hecho', 'descripcion_conducta'])['cantidad'].sum().unstack()

# Crear el gráfico de barras
delitos_por_año.plot(kind='bar', figsize=(10,6))

# Añadir títulos y etiquetas
plt.title('Cantidad total de víctimas por delito por cada año', fontsize=14)
plt.xlabel('Año', fontsize=12)
plt.ylabel('Cantidad de víctimas', fontsize=12)

# Mostrar el gráfico
plt.show()
```

```python
#Genera un gráfico de barras que muestre la cantidad de víctimas en los 5 departamentos con más casos reportados para el año 2024.
# Filtrar los datos para el año 2024
victimas_2024 = df_clean[df_clean['año_hecho'] == 2024]

# Agrupar por departamento y sumar las víctimas
top_5_departamentos = victimas_2024.groupby('departamento')['cantidad'].sum().nlargest(5)

# Crear el gráfico de barras
top_5_departamentos.plot(kind='bar', figsize=(10,6), color='c')

# Añadir títulos y etiquetas
plt.title('Cantidad de víctimas en los 5 departamentos con más casos reportados en 2024', fontsize=14)
plt.xlabel('Departamento', fontsize=12)
plt.ylabel('Cantidad de víctimas', fontsize=12)

# Mostrar el gráfico
plt.show()
```

```python
#Haz un gráfico de torta que muestre la proporción de casos de trata de personas frente a tráfico de migrantes del año 2024.
# Filtrar los casos de 2024
casos_2024 = df_clean[df_clean['año_hecho'] == 2024]

# Agrupar por delito y sumar
proporcion_casos = casos_2024.groupby('descripcion_conducta')['cantidad'].sum()

# Crear el gráfico de torta
plt.figure(figsize=(8,8))
plt.pie(proporcion_casos, labels=proporcion_casos.index, autopct='%1.1f%%', colors=['lightcoral', 'lightskyblue','yellow'])

# Añadir título
plt.title('Proporción de casos de trata de personas vs tráfico de migrantes en 2024', fontsize=14)

# Mostrar el gráfico
plt.show()
```

## Ejercicio 7: Identificación de picos en la cantidad de víctimas

1. Encuentra los meses o años con mayor cantidad de víctimas.
2. Crea un gráfico que muestre los picos de casos en cada año o mes, si es posible.




```python
#meses o años con mayor cantidad de víctimas.
# Extraer el mes de la columna 'fecha_hecho'
df_clean['mes_hecho'] = df_clean['fecha_hecho'].dt.month

# Agrupar los datos por año y mes para sumar la cantidad de víctimas
victimas_por_año_mes = df_clean.groupby(['año_hecho', 'mes_hecho'])['cantidad'].sum().reset_index()

# Encontrar los meses con mayor cantidad de víctimas
max_victimas_por_mes = victimas_por_año_mes.loc[victimas_por_año_mes['cantidad'].idxmax()]

# Mes y año con mayor cantidad de víctimas
print(f"El mes con mayor cantidad de víctimas es: mes {max_victimas_por_mes['mes_hecho']} del año {max_victimas_por_mes['año_hecho']} con {max_victimas_por_mes['cantidad']} víctimas.")
```

```python
# Agrupar los datos por año y mes para sumar la cantidad de víctimas
victimas_por_año_mes = df_clean.groupby(['año_hecho', 'mes_hecho'])['cantidad'].sum().reset_index()

# Encontrar el pico más alto de cada año (mes con más víctimas por año)
picos_por_año = victimas_por_año_mes.loc[victimas_por_año_mes.groupby('año_hecho')['cantidad'].idxmax()]

# Mostrar los resultados
print(picos_por_año)

# Crear el gráfico de barras para mostrar el pico más alto de cada año
plt.figure(figsize=(10,6))

# Crear un gráfico de barras con los picos de cada año
plt.bar(picos_por_año['año_hecho'].astype(str) + "-" + picos_por_año['mes_hecho'].astype(str), 
        picos_por_año['cantidad'], color='orange')

# Añadir títulos y etiquetas
plt.title('Pico más alto de víctimas por año y mes', fontsize=14)
plt.xlabel('Año-Mes', fontsize=12)
plt.ylabel('Cantidad de víctimas', fontsize=12)

# Mostrar el gráfico
plt.xticks(rotation=45)  # Rotar las etiquetas del eje X para mayor claridad
plt.show()
```

# 3 hallazgos importantes sobre la trata de personas y el tráfico de migrantes en Colombia.

A partir de los gráficos generados, estos son tres hallazgos clave sobre la trata de personas y el tráfico de migrantes en Colombia:

### 1. **Predominancia de la trata de personas sobre el tráfico de migrantes**
   En el gráfico de torta que muestra la proporción de casos en 2024, se observa que la trata de personas (artículo 188 A) constituye la vasta mayoría de los casos, representando aproximadamente el 91.5% de todos los delitos reportados. Esto resalta la gravedad y el enfoque necesario en este delito, mientras que el tráfico de migrantes (artículo 188) y otras circunstancias de agravación punitiva tienen una presencia mucho menor, indicando que la lucha contra la trata de personas debe tener mayor prioridad en términos de políticas y medidas preventivas.

### 2. **Bogotá D.C. como el epicentro de los casos en 2024**
   En el gráfico de barras que muestra los 5 departamentos con más casos reportados en 2024, Bogotá D.C. sobresale de manera significativa con más de 60 casos, mientras que otros departamentos como Antioquia, Nariño, Norte de Santander y Valle del Cauca reportan números considerablemente menores. Esto sugiere que la capital es el principal foco de estos delitos, lo cual puede deberse a su gran densidad poblacional, su rol como centro de movilidad y conexiones internacionales, y tal vez la facilidad para ocultar actividades ilícitas en una ciudad tan grande.

### 3. **Patrón de picos cíclicos en los casos de trata de personas y tráfico de migrantes**
   El gráfico que muestra los picos más altos de casos por mes y año revela que en algunos años, como 2008 y 2015, hubo incrementos significativos en la cantidad de víctimas. Esto sugiere que podrían existir factores contextuales o específicos de cada año que generan estas olas de delitos. Estos picos cíclicos podrían estar relacionados con factores socioeconómicos, cambios en la legislación, o actividades específicas de redes criminales, lo que indica la necesidad de monitorear estos cambios para anticiparse a futuros aumentos de estos reportados.


## Ejercicio 9: Análisis libre

1. Realiza un análisis adicional a partir de los datos. Puedes elegir una nueva pregunta o un patrón que te parezca interesante explorar. Algunas ideas podrían ser:
   - Analizar la evolución temporal de los casos.

```python
#Realizaremos la agrupación por zona geográfica y analizaremos patrones
#Para ello primero agrupamos los datos por departamento
# Agrupar los datos por departamento y sumar la cantidad de víctimas
victimas_por_departamento = df_clean.groupby('departamento')['cantidad'].sum().reset_index()

# Mostrar los departamentos con más casos
victimas_por_departamento
```

## ¿Por qué?
Agrupar por departamento permite ver la distribución geográfica de los delitos y entender en qué zonas se concentran más casos. Esto es fundamental para tomar decisiones sobre cómo y dónde desplegar recursos preventivos y de atención a las víctimas.

```python
# Ordenamos los datos por cantidad de víctimas
victimas_por_departamento_ordenado = victimas_por_departamento.sort_values(by='cantidad', ascending=False)

# Mostrar los departamentos ordenados
print(victimas_por_departamento_ordenado.head(10))  # Mostraremos los primeros 10 más afectados
```

```python
#Mediante un gráfico veremos de manera más sencilla la proporción de los datos según zona geografica
# Crear un gráfico de barras con los 10 departamentos más afectados
plt.figure(figsize=(10,6))
plt.bar(victimas_por_departamento_ordenado['departamento'].head(10), 
        victimas_por_departamento_ordenado['cantidad'].head(10), color='lightblue')

# Añadir títulos y etiquetas
plt.title('Cantidad de víctimas por departamento', fontsize=14)
plt.xlabel('Departamento', fontsize=12)
plt.ylabel('Cantidad de víctimas', fontsize=12)
plt.xticks(rotation=45)

# Mostrar el gráfico
plt.show()
```

```python
#Vamos a analizar los 5 departamentos más afectados a través del tiempo, los picos que encontremos en estos nos pueden dar indicios de factores externos 
# que influyeron

top_5_departamentos = victimas_por_departamento_ordenado.head(5)

# Agrupar los datos por año y ver cómo ha evolucionado la cantidad de víctimas en estos departamentos
evolucion_top_5 = df_clean[df_clean['departamento'].isin(top_5_departamentos['departamento'])] \
                  .groupby(['año_hecho', 'departamento'])['cantidad'].sum().unstack()

# Crear un gráfico de líneas para ver la evolución de casos en los 5 departamentos más afectados
evolucion_top_5.plot(kind='line', figsize=(12,6), marker='o')

# Añadir títulos y etiquetas
plt.title('Evolución de la cantidad de víctimas en los 5 departamentos más afectados', fontsize=14)
plt.xlabel('Año', fontsize=12)
plt.ylabel('Cantidad de víctimas', fontsize=12)
plt.grid(True)
plt.show()
```
